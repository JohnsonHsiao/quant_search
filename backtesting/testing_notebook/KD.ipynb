```python
import talib as ta
from backtesting import Backtest, Strategy
from backtesting.lib import crossover
import pandas as pd
import yfinance as yf
df = yf.download('BTC-USD ',start='2020-01-01')
df
```

```python
class KDCross(Strategy):
    
    def init(self):
        
        self.slowk, self.slowd = self.I(ta.STOCH, self.data.High, self.data.Low, 
                                        self.data.Close ,fastk_period=9, slowk_period=3, slowd_period=3)
    def next(self):
        if crossover(self.slowk, self.slowd):
            self.buy()
        elif crossover(self.slowd, self.slowk):
            self.sell()
            
bt = Backtest(df, KDCross, cash = 1000000, exclusive_orders=True)
output = bt.run()
output
```
